```python
import seaborn as sns
sns.set_theme(style="ticks")

dots = sns.load_dataset("dots")

# Define the palette as a list to specify exact values
palette = sns.color_palette("rocket_r")

# Plot the lines on two facets
sns.relplot(
    data=dots,
    x="time", y="firing_rate",
    hue="coherence", size="choice", col="align",
    kind="line", size_order=["T1", "T2"], palette=palette,
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
```

```python
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")

# load example tips dataset
tips=sns.load_dataset("tips")

# draw a nested boxplot to show bills by day and time
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)

sns.despine(offset=10, trim=True)
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# load exmple of tips dataset
tips=sns.load_dataset("tips")

# draw a nested violinplot to split the violins for easies comparision
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker",
              split=True, inner="quart", linewidth=1,
              palette={"Yes":"b", "No":".85"})

sns.despine(left=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

# load exmple of tips dataset
diamonds=sns.load_dataset("diamonds")

# draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset

f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price", hue="clarity", size="depth",
              palette="ch:r=-.2,d=.3_r",
              hue_order=clarity_ranking, sizes=(1, 8), linewidth=0,
              data=diamonds, ax=ax)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks")

# Initilize the figure with a loarathmic x-axis
f, ax = plt.subplots(figsize=(7, 6))
ax.set_xscale("log")

# load exmple of plamets dataset
planets=sns.load_dataset("planets")

# plot the orbital period with horizontal boxes
sns.boxplot(x="distance", y="method", data=planets, whis=[0,100],
           width=.6, palette="vlag")

# add in points to show each observation
sns.stripplot(x="distance", y="method", data=planets, size=4,
          color=".3", linewidth=0)

# tewak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```
